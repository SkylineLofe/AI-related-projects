#### Response

```python
def response(query,max_tokens=128,temperature=0,top_p=0.95,top_k=50):
    model_output = llm(
      prompt=query,
      max_tokens=max_tokens,
      temperature=temperature,
      top_p=top_p,
      top_k=top_k
    )

    clean_output = model_output['choices'][0]['text']
    clean_output = clean_output.replace('\n', ' ').replace('\n2', ' ').strip()
    return clean_output
```

### Data Chunking

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
text_chunks = text_splitter.split_documents(docs)

print(f"Total number of text chunks: {len(text_chunks)}")
```

### Embedding

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
embeddings = embedding_model.embed_documents([chunk.page_content for chunk in text_chunks])

print(f"Generated {len(embeddings)} embeddings.")
print(f"Shape of the first embedding: {np.array(embeddings[0]).shape}")
```

### Vector Database

```python
!pip install chromadb
```

```python
import chromadb
vectordb = Chroma.from_documents(
    documents=text_chunks,
    embedding=embedding_model,
    persist_directory="./chroma_db"
)
print("Vector database created and persisted successfully.")
```

### Retriever

```python
retriever = vectordb.as_retriever(search_kwargs={"k": 3})
```

### System and User Prompt Template

```python
qna_system_message = system_prompt
qna_user_message_template = """Context: {context}

Question: {question}

Answer:"""
```

### Response Function

```python
def generate_rag_response(user_input,k=3,max_tokens=128,temperature=0,top_p=0.95,top_k=50):
    global qna_system_message,qna_user_message_template
    # Retrieve relevant document chunks
    relevant_document_chunks = vectordb.similarity_search(query=user_input,k=k)
    context_list = [d.page_content for d in relevant_document_chunks]

    # Combine document chunks into a single context
    context_for_query = ". ".join(context_list)

    user_message = qna_user_message_template.replace('{context}', context_for_query)
    user_message = user_message.replace('{question}', user_input)

    prompt = qna_system_message + '\n' + user_message

    # Generate the response
    try:
        response = llm(
                  prompt=prompt,
                  max_tokens=max_tokens,
                  temperature=temperature,
                  top_p=top_p,
                  top_k=top_k
                  )

        # Extract and print the model's response
        # Replace literal '\n' and '\n2' sequences that appear in the model's output
        response = response['choices'][0]['text'].replace('\n', ' ').replace('\n2', ' ').strip()
    except Exception as e:
        response = f'Sorry, I encountered the following error: \n {e}'

    return response
```

## Output Evaluation

Let us now use the LLM-as-a-judge method to check the quality of the RAG system on two parameters - retrieval and generation. We illustrate this evaluation based on the answeres generated to the question from the previous section.

- We are using the same Mistral model for evaluation, so basically here the llm is rating itself on how well he has performed in the task.

```python
groundedness_rater_system_message  = """You are an impartial and helpful assistant. Your task is to evaluate the provided answer based *solely* on the given context. Do not use any outside knowledge. Rate the answer's groundedness on a scale of 1 to 5, where 1 means the answer completely contradicts the context or contains fabricated information, and 5 means the answer is fully supported by the context and contains no external information. Provide a brief explanation for your rating."""
```

```python
relevance_rater_system_message = """You are an impartial and helpful assistant. Your task is to evaluate how relevant the provided answer is to the question. The answer should directly and comprehensively address the question, utilizing the provided context as appropriate. Rate the answer's relevance on a scale of 1 to 5, where 1 means the answer is completely irrelevant to the question, and 5 means the answer is highly relevant and addresses all aspects of the question. Provide a brief explanation for your rating."""
```

```python
user_message_template = """Context: {context}\nQuestion: {question}\nAnswer: {answer}\nRating:"""
```

```python
def generate_ground_relevance_response(user_input,k=3,max_tokens=128,temperature=0,top_p=0.95,top_k=50):
    global qna_system_message,qna_user_message_template
    # Retrieve relevant document chunks
    relevant_document_chunks = vectordb.similarity_search(query=user_input,k=k)
    context_list = [d.page_content for d in relevant_document_chunks]
    context_for_query = ". ".join(context_list)

    # Combine user_prompt and system_message to create the prompt
    prompt = f"""[INST]{qna_system_message}
                {'user'}: {qna_user_message_template.format(context=context_for_query, question=user_input)}
                [/INST]"""

    response = llm(
            prompt=prompt,
            max_tokens=max_tokens,
            temperature=temperature,
            top_p=top_p,
            top_k=top_k,
            stop=['INST'],
            )

    answer =  response["choices"][0]["text"]

    # Combine user_prompt and system_message to create the prompt
    groundedness_prompt = f"""[INST]{groundedness_rater_system_message}
                {'user'}: {user_message_template.format(context=context_for_query, question=user_input, answer=answer)}
                [/INST]"""

    # Combine user_prompt and system_message to create the prompt
    relevance_prompt = f"""[INST]{relevance_rater_system_message}
                {'user'}: {user_message_template.format(context=context_for_query, question=user_input, answer=answer)}
                [/INST]"""

    response_1 = llm(
            prompt=groundedness_prompt,
            max_tokens=max_tokens,
            temperature=temperature,
            top_p=top_p,
            top_k=top_k,
            stop=['INST'],
            )

    response_2 = llm(
            prompt=relevance_prompt,
            max_tokens=max_tokens,
            temperature=temperature,
            top_p=top_p,
            top_k=top_k,
            stop=['INST'],
            )

    return response_1['choices'][0]['text'],response_2['choices'][0]['text']
```

```python
groundedness_rating, relevance_rating = generate_ground_relevance_response("What is the protocol for managing sepsis in a critical care unit?", k=3)
print("Groundedness Rating:", groundedness_rating)
print("Relevance Rating:", relevance_rating)
```
